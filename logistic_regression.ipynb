```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# creating a new dataset
from sklearn.datasets import make_classification
```

```python
# also make_classification function se data standarized form mai ata hy so we dont have to repeat that step
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)

# no of features = 10, means no of independent features (columns)
# no of classes, means how many categoreis, here we chose 2
#  X = independent feature
# Y = output feature
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=15)
```

```python
# model training
from sklearn.linear_model import LogisticRegression
# when penalty l1 l2 applied makes this logistic model into lasso or ridge regression accordingly, as logistic is type of linear model coz we finding a best fit line to divide datapooints into classes

# ridge is used for overfitting and lasso is used for feature selection 
# c is inversely propotional to lambda which is a imp factor in ridge and lasso regression
# class weight is used to balance imbalance dataset, we can decide weight of classes accordinlgy if theyre imbalanced


logistic = LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred = logistic.predict(X_test)
```

```python
# tells about our new data that what proabability it has to fall in diff classes, whichever class has more probability that means new point will fall on that category or give that output
# here we had two classes so its giving probability of each test point according to both classes


logistic.predict_proba(X_test)
```

```python
# thru classification report we get precison , recall and all
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score = accuracy_score(y_test,y_pred)
print(score)
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(classification_report(y_test,y_pred))
```

# hyperparameter tuning and cross validation


```python
# checking which is the best paramter for out model, thats why taking all possible parameters in list
model = LogisticRegression()
penalty = ['l1','l2','elasticnet']
```

```python
c_values = [100,10,0.1,0.01]
```

```python
solver = ['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
# these are all possible  values for parameters of logistic reg model, we want to see which value is best for our model
# then we will assign that value permanently to our model's paramter respectively for each parameter
params = dict(penalty=penalty,C=c_values,solver=solver)
```

```python
# gridsearchcv
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
# n = 5, shuffle = false, by default
```

```python
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
# njobs = -1 so that it uses all processors in our system
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
# the best parameters have been selected for our model, after trying all the combinations of given option, these are the best paramters
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
# thru classification report we get precison , recall and all
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score = accuracy_score(y_test,y_pred)
print("score: ",score)
cm = confusion_matrix(y_test,y_pred)
print("confusion matrix: ",cm)
print(classification_report(y_test,y_pred))
```

# Randomizes Search cv

grid search takes all the combinations of all the paramters provided, it will make time so in order to overcome this we use randomized search cv , which takes random combination and then finds out which combination suits best on our problem statement

The main difference is that in large no of dataset randomized search cv will take less time than grid search cv

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = LogisticRegression()
randomcv = RandomizedSearchCV(estimator = model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred = randomcv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score = accuracy_score(y_test,y_pred)
print("score: ",score)
cm = confusion_matrix(y_test,y_pred)
print("confusion matrix: ",cm)
print(classification_report(y_test,y_pred))
```

# Logistic Regression for multi-class classification

```python
from sklearn.datasets import make_classification
# when classes > 2 use n_informative paramter as well before that
X,y=make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=15)

```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
# give multi class paramter as ovr for multi class classification
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(multi_class='ovr')
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred = logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score = accuracy_score(y_test,y_pred)
print("score: ",score)
cm = confusion_matrix(y_test,y_pred)
print("confusion matrix: ",cm)
print(classification_report(y_test,y_pred))
```
