# Wordnet Lemmatizer
Lematization technique is like stemming. The output we will get after lematization is called 'lemma',which is a root word , rather than root stem, the output of stemming. After lemmatization,we will be getting valid word that means the same thing as the input word

```python
!pip install nltk
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lematizer = WordNetLemmatizer()
```

```python
""" 
POS (second paramter of lematizer, to know nature of word, by default its noun)
Noun - n
verb -v
adjective - a
adverb - r

"""
```

```python
import nltk
nltk.download('wordnet')
```

```python
print(lematizer.lemmatize("going",pos='n'))
print(lematizer.lemmatize("going",pos='v')) # here we can see when we give correct
#nature of 'going' that is verb, we get correct root word
```

```python
words = ["eating","eaten","eats","writing","writes","programming","program","history","finally","finalize"]
```

```python
for word in words:
    print(f"{word} -> {lematizer.lemmatize(word,pos='v')}")
```

```python
print(f"{lematizer.lemmatize("sportingly",pos='v')}")
```
