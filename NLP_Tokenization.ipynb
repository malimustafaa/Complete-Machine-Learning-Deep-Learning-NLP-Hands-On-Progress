```python
#corpus = paragraph

corpus = "My name is muhammad ali and im learning nlp from scratch. I hope by end of this course i'll become quite good with nlp! and further excell."
```

```python
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt')
```

```python
documents = sent_tokenize(corpus) #documents = sentences
```

```python
documents
```

```python
## Tokenization 
# paragraph -> words
#senence->words

from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
# use wordpunct to seperate punctuations too
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
#it seperate i'll -> i, ' , ll
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
new = tokenizer.tokenize(corpus) # doesnt treat . and ! or any other punctation as sepearte word, it is added with previous word,but with the last most word , full stop will be seperate as a new word
```
