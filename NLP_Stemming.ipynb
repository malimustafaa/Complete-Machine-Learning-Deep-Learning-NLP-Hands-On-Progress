# Stemming 
stemming is the process of reducing a word to its word stem. eg:

goes,going,gone -> go

eating,eats,eaten -> eat

```python
words = ["eating","eaten","eats","writing","writes","programming","program","history","finally","finalize"]
```

# PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(f"{word} ----> {stemming.stem(word)}")
```

disadvantage of stemming is for some words after stemming,the entire meanin changes like here history becomes histori, which is incorrect

This disadvantage gets fixed with lemmatization

```python
#another example of disadvantage
stemming.stem('congratulations')
```

# Regex Stemmer class

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stem = RegexpStemmer('$ing|s$|e$|able$', min=4) # ing$ means word ending with
#ing , from that word ing and will be removed,just the ing will be removed.
#if i pass 'ing' as an argument istead of ing$ then all the ings will be removed from the word whether its prior to word or after in

```

```python
reg_stem.stem("eatingsw")
```

```python
reg_stem.stem('ingstemmering') # here i changed parameter to 'ing' from '$ing'
```

# snowball stemmer 
snowball stemmer performs better than PorterStemmer by giving better form of a word, and less meaningless words 

```python
from nltk.stem import SnowballStemmer
```

```python
snow = SnowballStemmer("english")
```

```python
for word in words:
    print(f" {word}-> {snow.stem(word)}")
```
