```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y = make_moons(n_samples=250,noise=0.05)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
# standard scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(X)
```

```python
x_scaled
```

```python
dbscan = DBSCAN(eps=0.3)
```

```python
dbscan.fit(x_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_) # with dbscan
```

```python
plt.scatter(X[:,0],X[:,1],c=y) # without dbscan, we can see k dbscan ne blkl theek cluster bnaye coz we know y in this case and y aka target value bh usi trha k cluster bna rha here, whichfurther confirms k sahi cluster bnaye
```
