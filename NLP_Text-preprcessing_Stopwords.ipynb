```python
## apply stop words and filter and then apply stemming

for i in range(len(sentence)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english')) ]
    sentence[i] = "  ".join(words) #joining the "words" list all elements together n to a sentence index [i] again with space in between , to put cleaned sentence back to its place
    
```
