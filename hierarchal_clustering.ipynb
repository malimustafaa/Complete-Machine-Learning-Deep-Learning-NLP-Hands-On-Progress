```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
```

```python
# import iris dataet
iris =datasets.load_iris()
```

```python
iris
```

```python
iris.keys()
```

```python
iris.feature_names
```

```python
iris_data = pd.DataFrame(iris.data,columns = iris['feature_names'])
```

```python
iris_data.head()
```

```python
print(iris.DESCR)
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
iris_scaled = scaler.fit_transform(iris_data)
```

```python
iris_scaled
```

```python
# applying pca
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
data_pca = pca.fit_transform(iris_scaled)
```

```python
data_pca
```

```python
plt.scatter(data_pca[:,0],data_pca[:,1],c=iris.target)
```

```python
# agglomerative clustering
 # to construct a dendogram
import scipy.cluster.hierarchy as sc
# plot the dendogran
plt.figure(figsize=(20,7))
plt.title("dendogram")
# create dendrogram

sc.dendrogram(sc.linkage(data_pca,method='ward')) #  different methods hein go and check agglomerative clusterin web for more methods and see what diff methods do
plt.xlabel("sample index")
plt.ylabel("euclidena distance") # by default hota ye k y axis pr eucliean distance hi hoga
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward') # n cluster 2 coz we clearly can see in dendrogram k  highest vertical line jisme koi horizontal pass nh hori, and uspe 2 hi point of crossing hori 
cluster.fit(data_pca)
```

```python
cluster.labels_
```

```python
plt.scatter(data_pca[:,0],data_pca[:,1],c=cluster.labels_)
# here we can see 2 clusters are present as compred to 3 clusters in our real data which we plotted above, 
# coz after aglo clustering and choosing highest verstical line we get k=2 coz we get 2 clusters as it is best suitable acc to the data
```
