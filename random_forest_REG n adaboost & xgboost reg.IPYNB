```python
num_features = X.select_dtypes(exclude='object').columns
```

```python
onehot_features = ['seller_type','fuel_type','transmission_type']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
 
numeric_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop='first')

pre_processor = ColumnTransformer(
    [
        ("OneHotEncoder",oh_transformer,onehot_features),
        ("StandardScaler",numeric_transformer,num_features)
    ],remainder='passthrough'
)

# remainder = passthrough means all other columns will stay as it is, if we dont use passthrough, it will delete all rest features
```

```python
# to see rows and cols
df.shape
```

```python
X = pre_processor.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train.shape,X_test.shape
```

### Model training and model selection

```python
from xgboost import XGBRegressor
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
# function to evaluate model
def evaluate_model(true,predicted):
    mae = mean_absolute_error(true,predicted)
    mse = mean_squared_error(true,predicted)
    rmse = np.sqrt(mean_squared_error(true,predicted))
    r2_square = r2_score(true,predicted)
    return mae,rmse,r2_square
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Lasso": Lasso(),
    "Ridge": Ridge(),
    "K-Neighbors Regressor": KNeighborsRegressor(),
    "DecisionTree": DecisionTreeRegressor(),
    "RandomForestRegressor": RandomForestRegressor(),
    "AdaBoostRegressor": AdaBoostRegressor(),
    "XG Boost Regresssor": XGBRegressor()
}
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train) # train the model

    # make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Evaluate train and test dataset

    model_train_mae, model_train_rmse, model_train_r2 = evaluate_model(y_train,y_train_pred)
    model_test_mae, model_test_rmse,model_test_r2 = evaluate_model(y_test,y_test_pred)

    print(list(models.keys())[i])
    print("Model performance for training set")
    print("Root mean sq error", round(model_train_rmse,2))
    print("mean absolute error",round(model_train_mae,2))
    print("R2 score",round(model_train_r2,2))
    print("----------------------------------------------------")
    print("\n")
    print("Model performance for test dataset")
   
    print("Root mean sq error", round(model_test_rmse,2))
    print("mean absolute error",round(model_test_mae,2))
    print("R2 score",round(model_test_r2,2))
    print('='*35)
    print('\n')



```

```python
# parameters for hyperpara tuning
#knn_params = {"n_neighbors": [2,3,10,20,40,50]}
rff_params = {"max_depth": [5,8,15,None,10],
              "max_features": [5,7,"auto",8],
              "min_samples_split": [2,8,15,20],
              "n_estimators": [100,200,500,1000]}
#adaboost_params = {"n_estimators": [50,60,70,80],
   #                "loss": ['linear','square','exponential']
    #               }
xg_param = { "learning_rate": [0.1,0.01],
            "max_depth": [5,8,12,20,30],
            "n_estimators": [100,200,300],
            "colsample_bytree": [0.5,0.8,1,0.3,0.4]

}


```

```python
randomcv_models = [
                   ('RFF',RandomForestRegressor(),rff_params),
                   ('XGG',XGBRegressor(),xg_param)
]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model_param = {}
for name,model,params in randomcv_models:
    random = RandomizedSearchCV(estimator = model,param_distributions=params,n_iter=100,cv=3,verbose=2,n_jobs=-1)
    random.fit(X_train,y_train)
    model_param[name] = random.best_params_
    
```

```python
model_param
```

```python
# Retraining of models wth best parameters
models = {
"Random Forest Regressor": RandomForestRegressor(n_estimators= 200,
  min_samples_split= 2,
  max_features= 7,
  max_depth= 15),
#"KNeighborsRegressor": KNeighborsRegressor(n_neighbors=2),
#"adaboostAlgo": AdaBoostRegressor(n_estimators= 60, loss= 'linear')
"XG_BOOST": XGBRegressor(n_estimators=300,max_depth=5,learning_rate=0.1,colsample_bytree=1)
}
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train) # train the model

    # make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Evaluate train and test dataset

    model_train_mae, model_train_rmse, model_train_r2 = evaluate_model(y_train,y_train_pred)
    model_test_mae, model_test_rmse,model_test_r2 = evaluate_model(y_test,y_test_pred)

    print(list(models.keys())[i])
    print("Model performance for training set")
    print("Root mean sq error", round(model_train_rmse,2))
    print("mean absolute error",round(model_train_mae,2))
    print("R2 score",round(model_train_r2,2))
    print("----------------------------------------------------")
    print("\n")
    print("Model performance for test dataset")
    print("Model performance for training set")
    print("Root mean sq error", round(model_test_rmse,2))
    print("mean absolute error",round(model_test_mae,2))
    print("R2 score",round(model_test_r2,2))
    print('='*35)
    print('\n')
```
