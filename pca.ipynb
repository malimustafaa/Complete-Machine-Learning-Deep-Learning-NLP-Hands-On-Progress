```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_data = load_breast_cancer()
```

```python
cancer_data.keys()
```

```python
print(cancer_data.DESCR)
```

```python
df = pd.DataFrame(cancer_data['data'],columns = cancer_data['feature_names'] )
```

```python
df.head()
```

```python
#standardization

# its a good practise beofre feature extraction (dimensionality reduction)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
# now the data has become standardized normal form , mean = 0, and standard deviation = 1
scaled_data
```

```python
# applying pca algorithms
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) # n_components mtlb kitne features mai convert krna like here we converting all these features to 2
```

```python
data_pca = pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pca.explained_variance_ # variance of 2 columns which are selected or all the cols are transformed into  
```

```python
# if we will not give n_component and run it without it, then wo utne hi cols bnayega jitne features thay
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_data['target'],cmap='plasma')
plt.xlabel('first principal compoent')
plt.ylabel('second principal component')
```
