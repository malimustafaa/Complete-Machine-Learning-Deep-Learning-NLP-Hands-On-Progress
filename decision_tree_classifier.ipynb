```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

splitting dependent and independent feaures

```python
X = pd.DataFrame(iris.data,columns = ['sepal length','sepal width','petal length','petal width'])
```

```python
X
```

```python
y = pd.DataFrame(iris.target)
```

```python
y.tail
```

train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_cl = DecisionTreeClassifier()
```

```python
tree_cl.fit(X_train,y_train)
```

```python
#visualize decision tree
from sklearn import tree
plt.figure(figsize=(15,20))
tree.plot_tree(tree_cl,filled=True)
```

post pruning to prevent overfitting

```python
# set max_depth = 2, as we cn see after 2ns split further splits are not required(start from 0 as root node)
# after 2nd split left side will give output as 2nd category coz 33 in 2nd position, and right side will give outpt as 3rd
# coz other position numbers are so small that further split isnt required. Leaf node as 1st split will give output as 1st category coz 40 on 1st place and all other positions are 0

tree_cl = DecisionTreeClassifier(max_depth=2)
```

```python
tree_cl.fit(X_train,y_train)
```

```python
#visualize decision tree
from sklearn import tree
plt.figure(figsize=(15,20))
tree.plot_tree(tree_cl,filled=True)
```

```python
y_pred = tree_cl.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
cm = confusion_matrix(y_test,y_pred)

print(cm)
cr = classification_report(y_test,y_pred)
print(cr)
```

# decision tree pre pruning and hyperparameter tuning

```python
param = {
    'criterion' : ['gini', 'entropy', 'log_loss'], 
    'splitter' : ['best','random'],
    'max_depth' : [1,2,3,4,5],
    'max_features': ['auto','sqrt','log2']
    

}
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(tree_model,param_grid=param,cv=5)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
cm = confusion_matrix(y_test,y_pred)

print(cm)
cr = classification_report(y_test,y_pred)
print(cr)
```
